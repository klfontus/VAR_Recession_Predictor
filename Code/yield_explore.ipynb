```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
y_df = pd.read_csv('yield_curve_T10Y2Y.csv') #data is from FRED website, all the yield curve data they have
y_df.columns = ["date", "yield"]
display(y_df.shape)
display(y_df.head())
```

```python
y_df = y_df[y_df["yield"]!='.']
#dropping rows that have a value of '.'
display(y_df.shape)
```

```python
y_df["date"] = pd.to_datetime(y_df["date"])
y_df["yield"] = pd.to_numeric(y_df["yield"])
y_df.info()
```

```python
plt.figure(figsize=(20,5))
plt.plot(y_df['date'], y_df['yield'], linestyle='-', label="Yield Curve")
plt.ylabel("(10-Yr Treasury Constant Maturity) - (2-Yr Treasury Constant Maturity)")
plt.xlabel("Date")

#plt.axvline(x=pd.to_datetime("1980-01-01"), color='r')
#plt.axvline(x=pd.to_datetime("1980-07-01"), color='r')

plt.axvspan(pd.to_datetime("1980-01-01"), pd.to_datetime("1980-07-01"), alpha=0.3, color='red', label="Recession")
plt.axvspan(pd.to_datetime("1981-07-01"), pd.to_datetime("1982-11-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("1990-07-01"), pd.to_datetime("1991-03-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("2001-03-01"), pd.to_datetime("2001-11-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("2007-12-01"), pd.to_datetime("2009-06-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("2020-02-01"), pd.to_datetime("2020-04-01"), alpha=0.3, color='red')
plt.legend()
```
