```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
gdp_df = pd.read_csv('Real_GDP_GDPC1.csv') #data is from FRED website????
gdp_df.columns = ["date", "gdp"]
display(gdp_df.shape)
display(gdp_df.head())
display(gdp_df.info())
```

```python
gdp_df["date"] = pd.to_datetime(gdp_df["date"])
display(gdp_df.info())
```

```python
plt.figure(figsize=(20,5))
plt.plot(gdp_df['date'], gdp_df['gdp'], linestyle='-', label="Real GDP (in billions of $USD)")
plt.ylabel("Real GDP")
plt.xlabel("Date")

#plt.axvline(x=pd.to_datetime("1980-01-01"), color='r')
#plt.axvline(x=pd.to_datetime("1980-07-01"), color='r')

plt.axvspan(pd.to_datetime("1948-11-01"), pd.to_datetime("1949-10-01"), alpha=0.3, color='red', label="Recession")
plt.axvspan(pd.to_datetime("1953-07-01"), pd.to_datetime("1954-05-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("1957-08-01"), pd.to_datetime("1958-04-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("1960-04-01"), pd.to_datetime("1961-02-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("1969-12-01"), pd.to_datetime("1970-11-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("1973-11-01"), pd.to_datetime("1975-03-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("1980-01-01"), pd.to_datetime("1980-07-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("1981-07-01"), pd.to_datetime("1982-11-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("1990-07-01"), pd.to_datetime("1991-03-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("2001-03-01"), pd.to_datetime("2001-11-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("2007-12-01"), pd.to_datetime("2009-06-01"), alpha=0.3, color='red')
plt.axvspan(pd.to_datetime("2020-02-01"), pd.to_datetime("2020-04-01"), alpha=0.3, color='red')
plt.legend()
```
