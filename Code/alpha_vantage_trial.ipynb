```python
def get_api_key(API_key_location):
    with open(API_key_location) as f:
        API_key = f.readlines()[0] # get the API key
        print('API Key: ' + API_key)
    return API_key

API_key_location = r"C:\Users\Edmund\Alpha Vantage\api key.txt" # Replace this with the path to your Alpha Vantage API key
API_key = get_api_key(API_key_location)
```

```python
from alpha_vantage.timeseries import TimeSeries
ts = TimeSeries(key=API_key, output_format='pandas', indexing_type='date') # configure output type for time series data
# Get data and metadata as pandas dataframe
data, meta_data = ts.get_monthly('MSFT')
display(data.head())
display(data.tail())
```
