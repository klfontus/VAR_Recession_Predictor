```python
# testing yfinance API. Look here for documentation: https://pypi.org/project/yfinance/ , https://analyzingalpha.com/yfinance-python
import pandas as pd

import yfinance as yf
msft = yf.Ticker("MSFT")
# get all stock info
mstf_info = msft.info
print(type(mstf_info))
print(len(mstf_info))
print(mstf_info)

msft_price = msft.history(interval='1d', start = '2004-06-03')
display(msft_price.head(12))
display(msft_price.tail(12))
print(msft_price.shape)
print(type(msft_price))
msft_price['Date'] = msft_price.index
msft_price.plot(x = 'Date', y ='Open', kind = 'scatter')
msft_price[msft_price['Dividends'] > 0].plot(x = 'Date', y ='Dividends', kind = 'scatter')
```
